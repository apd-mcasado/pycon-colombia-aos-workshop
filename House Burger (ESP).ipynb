## üéØ Funci√≥n Objetivo

Minimizar el costo total esperado:
- Ingredientes utilizados en la cocina
- Entregas realizadas
- Almacenamiento utilizado
- Valor esperado de la penalizaci√≥n por demanda no satisfecha

$$
\min 
\sum_{t \in T}\sum_{p \in P}\sum_{i \in I} c_i  q_{ip}  x_{pt} +
\sum_{t \in T} k  z_t +
\sum_{t \in T}\sum_{i \in I} g s_{it} +
\sum_{e \in E}\sum_{t \in T}\sum_{p \in P} \lambda \pi_u u_{pte}
$$

---

## ‚úÖ Restricciones


### 1. Demanda M√≠nima

$$
x_{pt} + u_{pte} \geq d_{pte} \quad \forall p \in P,\ t \in T,\ e \in E
$$


### 2. Din√°mica del Inventario

Para el primer d√≠a (<em>t = 1</em>):

$$
s_{i1} = s_{i0} - \sum_{p \in P} q_{ip} \cdot x_{p1} \quad \forall i \in I
$$

Para los d√≠as restantes (<em>t > 1</em>), considerando 2 d√≠as de retraso en la entrega:

$$
s_{it} = s_{i(t-1)} - \sum_{p \in P} q_{ip} \cdot x_{pt} + y_{i(t-2)} \quad \forall i \in I,\ t = 2,3,\dots,7
$$


### 3. Capacidad del Almac√©n

$$
\sum_{i \in I} v_i s_{it} \leq w \quad \forall t \in T
$$


### 4. Activaci√≥n de Entrega (Big-M)

$$
\sum_{i \in I} y_{it} \leq M \cdot z_t \quad \forall t \in T
$$

Donde <em>M</em> es una constante suficientemente grande.

---


## üîí Resumen de Dominios de Variables


**Ventas de Productos**

$$

$$


**Cantidad de Ingredientes Pedidos**

$$
y_{it} \in \mathbb{Z}^+ \quad \forall i \in I,\ t \in T
$$


**Activaci√≥n de Entrega (Binaria)**

$$
z_t \in \lbrace0, 1\rbrace \quad \forall t \in T
$$


**Inventario de Ingredientes**

$$
s_{it} \in \mathbb{R}^+ \quad \forall i \in I,\ t \in T
$$


**Demanda no satisfecha (holgura)**

$$
u_{pte} \in \mathbb{R}^+ \quad \forall p \in P,\ t \in T, e \in E 
$$

```python
import pyomo.environ as pyo
from pyomo.opt import SolverStatus, TerminationCondition
import pandas as pd
import numpy as np

def resolver_optimizacion_house_burger():
    """
    Resolver el problema de optimizaci√≥n semanal de House Burger usando Pyomo
    """
    
    # Crear el modelo
    modelo = pyo.ConcreteModel()
    
    # =============================================================================
    # CONJUNTOS
    # =============================================================================
    
    # Conjunto de Productos
    modelo.P = pyo.Set(initialize=['BB', 'BC', 'CB', 'CC'])
    
    # Conjunto de Ingredientes
    modelo.I = pyo.Set(initialize=['Panes', 'Carne', 'Pollo', 'Papas', 'Aceite'])
    
    # Conjunto de D√≠as (1=Lunes, 2=Martes, ..., 7=Domingo)
    modelo.T = pyo.Set(initialize=range(1, 8))

    # Conjunto de Escenarios de Demanda
    modelo.E = pyo.Set(initialize=['Optimista', 'Esperado', 'Pesimista'])

    
    # =============================================================================
    # PAR√ÅMETROS
    # =============================================================================
    
    # Costo de env√≠o por pedido
    modelo.k = pyo.Param(initialize=25, mutable=True)
    
    # Capacidad del almac√©n
    modelo.w = pyo.Param(initialize=350)

    # Tarifa de almacenamiento por litro de ingrediente por noche
    modelo.f = pyo.Param(initialize=0.1)

    # Penalizaci√≥n por unidad de demanda no satisfecha
    modelo.lambda_penalizacion = pyo.Param(initialize=5)

    # Costo por unidad de ingrediente
    costos_ingredientes = {
        'Panes': 0.50,
        'Carne': 1.80,
        'Pollo': 1.50,
        'Papas': 0.70,
        'Aceite': 10.00
    }
    modelo.c = pyo.Param(modelo.I, initialize=costos_ingredientes)

    volumen_ingredientes = {
        'Panes': 0.32,
        'Carne': 0.27,
        'Pollo': 0.5,
        'Papas': 0.4,
        'Aceite': 4.0
    }
    modelo.v = pyo.Param(modelo.I, initialize=volumen_ingredientes)
    
    # Uso de ingredientes por producto
    uso_ingredientes = {
        ('Panes', 'BB'): 1, ('Carne', 'BB'): 1, ('Pollo', 'BB'): 0, ('Papas', 'BB'): 0, ('Aceite', 'BB'): 0,
        ('Panes', 'BC'): 1, ('Carne', 'BC'): 1, ('Pollo', 'BC'): 0, ('Papas', 'BC'): 1, ('Aceite', 'BC'): 0.1,
        ('Panes', 'CB'): 1, ('Carne', 'CB'): 0, ('Pollo', 'CB'): 1, ('Papas', 'CB'): 0, ('Aceite', 'CB'): 0.1,
        ('Panes', 'CC'): 1, ('Carne', 'CC'): 0, ('Pollo', 'CC'): 1, ('Papas', 'CC'): 1, ('Aceite', 'CC'): 0.2
    }
    modelo.q = pyo.Param(modelo.I, modelo.P, initialize=uso_ingredientes)

    # Probabilidades de que ocurra cada escenario de demanda
    probabilidades_escenarios = {
        'Pesimista': 0.2,
        'Esperado': 0.5,
        'Optimista': 0.3
    }
    modelo.pi = pyo.Param(modelo.E, initialize=probabilidades_escenarios)


    # Demanda m√≠nima por producto por d√≠a
    datos_demanda = {
        # Lunes
        ('BB', 1, 'Pesimista'): 24, ('BB', 1, 'Esperado'): 30, ('BB', 1, 'Optimista'): 36,
        ('BC', 1, 'Pesimista'): 32, ('BC', 1, 'Esperado'): 40, ('BC', 1, 'Optimista'): 48,
        ('CB', 1, 'Pesimista'): 16, ('CB', 1, 'Esperado'): 20, ('CB', 1, 'Optimista'): 24,
        ('CC', 1, 'Pesimista'): 28, ('CC', 1, 'Esperado'): 35, ('CC', 1, 'Optimista'): 42,

        # Martes
        ('BB', 2, 'Pesimista'): 29, ('BB', 2, 'Esperado'): 36, ('BB', 2, 'Optimista'): 43,
        ('BC', 2, 'Pesimista'): 30, ('BC', 2, 'Esperado'): 37, ('BC', 2, 'Optimista'): 44,
        ('CB', 2, 'Pesimista'): 20, ('CB', 2, 'Esperado'): 25, ('CB', 2, 'Optimista'): 30,
        ('CC', 2, 'Pesimista'): 26, ('CC', 2, 'Esperado'): 33, ('CC', 2, 'Optimista'): 40,

        # Mi√©rcoles
        ('BB', 3, 'Pesimista'): 28, ('BB', 3, 'Esperado'): 35, ('BB', 3, 'Optimista'): 42,
        ('BC', 3, 'Pesimista'): 32, ('BC', 3, 'Esperado'): 40, ('BC', 3, 'Optimista'): 48,
        ('CB', 3, 'Pesimista'): 12, ('CB', 3, 'Esperado'): 15, ('CB', 3, 'Optimista'): 18,
        ('CC', 3, 'Pesimista'): 33, ('CC', 3, 'Esperado'): 41, ('CC', 3, 'Optimista'): 49,

        # Jueves
        ('BB', 4, 'Pesimista'): 26, ('BB', 4, 'Esperado'): 33, ('BB', 4, 'Optimista'): 40,
        ('BC', 4, 'Pesimista'): 34, ('BC', 4, 'Esperado'): 43, ('BC', 4, 'Optimista'): 52,
        ('CB', 4, 'Pesimista'): 13, ('CB', 4, 'Esperado'): 16, ('CB', 4, 'Optimista'): 19,
        ('CC', 4, 'Pesimista'): 27, ('CC', 4, 'Esperado'): 34, ('CC', 4, 'Optimista'): 41,

        # Viernes
        ('BB', 5, 'Pesimista'): 45, ('BB', 5, 'Esperado'): 56, ('BB', 5, 'Optimista'): 67,
        ('BC', 5, 'Pesimista'): 50, ('BC', 5, 'Esperado'): 62, ('BC', 5, 'Optimista'): 74,
        ('CB', 5, 'Pesimista'): 40, ('CB', 5, 'Esperado'): 50, ('CB', 5, 'Optimista'): 60,
        ('CC', 5, 'Pesimista'): 40, ('CC', 5, 'Esperado'): 50, ('CC', 5, 'Optimista'): 60,

        # S√°bado
        ('BB', 6, 'Pesimista'): 30, ('BB', 6, 'Esperado'): 38, ('BB', 6, 'Optimista'): 46,
        ('BC', 6, 'Pesimista'): 44, ('BC', 6, 'Esperado'): 55, ('BC', 6, 'Optimista'): 66,
        ('CB', 6, 'Pesimista'): 28, ('CB', 6, 'Esperado'): 35, ('CB', 6, 'Optimista'): 42,
        ('CC', 6, 'Pesimista'): 34, ('CC', 6, 'Esperado'): 43, ('CC', 6, 'Optimista'): 52,

        # Domingo
        ('BB', 7, 'Pesimista'): 34, ('BB', 7, 'Esperado'): 42, ('BB', 7, 'Optimista'): 50,
        ('BC', 7, 'Pesimista'): 44, ('BC', 7, 'Esperado'): 55, ('BC', 7, 'Optimista'): 66,
        ('CB', 7, 'Pesimista'): 30, ('CB', 7, 'Esperado'): 37, ('CB', 7, 'Optimista'): 44,
        ('CC', 7, 'Pesimista'): 30, ('CC', 7, 'Esperado'): 37, ('CC', 7, 'Optimista'): 44,
    }
    modelo.d = pyo.Param(modelo.P, modelo.T, modelo.E, initialize=datos_demanda)
    
    # Inventario inicial
    inventario_inicial = {
        'Panes': 550,
        'Carne': 150,
        'Pollo': 150,
        'Papas': 200,
        'Aceite': 5
    }
    modelo.s0 = pyo.Param(modelo.I, initialize=inventario_inicial)
    
    # Par√°metro Big M para la restricci√≥n de activaci√≥n de entrega
    modelo.M = pyo.Param(initialize=100000)
    
    # =============================================================================
    # VARIABLES DE DECISI√ìN
    # =============================================================================
    
    # N√∫mero de productos a vender
    modelo.x = pyo.Var(modelo.P, modelo.T, domain=pyo.NonNegativeIntegers)
    
    # Cantidad de ingredientes a ordenar
    modelo.y = pyo.Var(modelo.I, modelo.T, domain=pyo.NonNegativeIntegers)
    
    # Activaci√≥n de entrega (binaria)
    modelo.z = pyo.Var(modelo.T, domain=pyo.Binary)
    
    # Inventario disponible al final de cada d√≠a
    modelo.s = pyo.Var(modelo.I, modelo.T, domain=pyo.NonNegativeReals)

    # Variables de holgura para demanda no satisfecha
    modelo.u = pyo.Var(modelo.P, modelo.T, modelo.E, domain=pyo.NonNegativeReals)

    # =============================================================================
    # FUNCI√ìN OBJETIVO
    # =============================================================================
    
    def regla_objetivo(modelo):
        # Costos de ingredientes usados en cocina
        costo_ingredientes = sum(modelo.c[i] * modelo.q[i, p]  * modelo.x[p, t] 
                               for i in modelo.I for t in modelo.T for p in modelo.P)

        # Costos de entrega
        costo_entrega = sum(modelo.k * modelo.z[t] for t in modelo.T)

        # Costos de almacenamiento
        costo_almacenamiento = sum(modelo.f * modelo.v[i] * modelo.s[i,t] 
                                  for i in modelo.I for t in modelo.T)

        # Valor esperado del costo por demanda no satisfecha
        costo_demanda_no_satisfecha_ve = sum(modelo.lambda_penalizacion * modelo.pi[e] * modelo.u[p,t,e] 
                                            for p in modelo.P for t in modelo.T for e in modelo.E)

        return costo_ingredientes + costo_entrega + costo_almacenamiento + costo_demanda_no_satisfecha_ve
    
    modelo.obj = pyo.Objective(rule=regla_objetivo, sense=pyo.minimize)
    
    # =============================================================================
    # RESTRICCIONES
    # =============================================================================
    
    # 1. Restricci√≥n de Demanda M√≠nima
    def regla_restriccion_demanda(modelo, p, t, e):
        return modelo.x[p, t] + modelo.u[p, t, e] >= modelo.d[p, t, e]
    
    modelo.restriccion_demanda = pyo.Constraint(modelo.P, modelo.T, modelo.E, rule=regla_restriccion_demanda)
    
    # 2. Din√°mica del Inventario
    # Para el d√≠a 1 (a√∫n no llegan entregas)
    def regla_inventario_dia1(modelo, i):
        return modelo.s[i, 1] == modelo.s0[i] - sum(modelo.q[i, p] * modelo.x[p, 1] for p in modelo.P)
    
    modelo.inventario_dia1 = pyo.Constraint(modelo.I, rule=regla_inventario_dia1)
    
    # Para el d√≠a 2 (a√∫n no llegan entregas)
    def regla_inventario_dia2(modelo, i):
        return modelo.s[i, 2] == modelo.s[i, 1] - sum(modelo.q[i, p] * modelo.x[p, 2] for p in modelo.P)
    
    modelo.inventario_dia2 = pyo.Constraint(modelo.I, rule=regla_inventario_dia2)
    
    # Para los d√≠as 3-7 (llegan entregas pedidas el d√≠a t-2)
    def regla_inventario_dias_restantes(modelo, i, t):
        if t >= 3:
            return modelo.s[i, t] == modelo.s[i, t-1] - sum(modelo.q[i, p] * modelo.x[p, t] for p in modelo.P) + modelo.y[i, t-2]
        else:
            return pyo.Constraint.Skip
    
    modelo.inventario_restante = pyo.Constraint(modelo.I, modelo.T, rule=regla_inventario_dias_restantes)
    
    # 3. Restricci√≥n de Capacidad del Almac√©n
    def regla_capacidad_almacen(modelo, t):
        return sum(modelo.v[i] * modelo.s[i, t] for i in modelo.I) <= modelo.w
    
    modelo.capacidad_almacen = pyo.Constraint(modelo.T, rule=regla_capacidad_almacen)
    
    # 4. Restricci√≥n de Activaci√≥n de Entrega (Big-M)
    def regla_activacion_entrega(modelo, t):
        return sum(modelo.y[i, t] for i in modelo.I) <= modelo.M * modelo.z[t]
    
    modelo.activacion_entrega = pyo.Constraint(modelo.T, rule=regla_activacion_entrega)
        
    # =============================================================================
    # RESOLVER EL MODELO
    # =============================================================================
    
    # Crear instancia del solver usando gurobi como optimizador
    solver = pyo.SolverFactory("gurobi", solver_io="python")
    
    
    print("üîÑ Resolviendo el Problema de Optimizaci√≥n de House Burger...")
    print("=" * 60)
    
    # Resolver el modelo
    resultados = solver.solve(modelo, tee=True)  # tee=True muestra salida del solver
    
    # Verificar si la soluci√≥n es √≥ptima
    if (resultados.solver.status == SolverStatus.ok) and \
       (resultados.solver.termination_condition == TerminationCondition.optimal):
        
        print("\n‚úÖ ¬°SOLUCI√ìN √ìPTIMA ENCONTRADA!")
        print("=" * 60)
        
        # Mostrar resultados
        mostrar_resultados(modelo)
        
        return modelo, resultados
    
    elif resultados.solver.termination_condition == TerminationCondition.infeasible:
        print("\n‚ùå ¬°EL PROBLEMA ES INFACTIBLE!")
        print("El problema no tiene soluci√≥n factible. Verifica las restricciones.")
        return None, resultados
    
    else:
        print(f"\n‚ö†Ô∏è  ESTADO DEL SOLVER: {resultados.solver.status}")
        print(f"CONDICI√ìN DE TERMINACI√ìN: {resultados.solver.termination_condition}")
        return modelo, resultados

def mostrar_resultados(modelo):
    """
    Mostrar los resultados de la optimizaci√≥n de forma formateada
    """
    
    print(f"üí∞ COSTO TOTAL M√çNIMO: ${pyo.value(modelo.obj):.2f}")
    print("=" * 60)
    
    # Cronograma de ventas de productos
    print("\nüìä CRONOGRAMA DIARIO DE VENTAS DE PRODUCTOS:")
    print("-" * 60)
    
    dias = ['Lunes', 'Martes', 'Mi√©rcoles', 'Jueves', 'Viernes', 'S√°bado', 'Domingo']
    
    datos_ventas = []
    for t in modelo.T:
        fila = {'D√≠a': dias[t-1]}
        for p in modelo.P:
            fila[p] = int(pyo.value(modelo.x[p, t]))
        datos_ventas.append(fila)
    
    df_ventas = pd.DataFrame(datos_ventas)
    print(df_ventas.to_string(index=False))
    
    # Cronograma de entregas
    print("\nüöö CRONOGRAMA DE PEDIDOS A REALIZAR:")
    print("-" * 60)
    
    datos_entregas = []
    for t in modelo.T:
        if pyo.value(modelo.z[t]) > 0.5:  # Variable binaria es 1
            pedidos_dia = {}
            pedido_total = 0
            for i in modelo.I:
                cantidad = int(pyo.value(modelo.y[i, t]))
                if cantidad > 0:
                    pedidos_dia[i] = cantidad
                    pedido_total += cantidad
            
            if pedido_total > 0:
                datos_entregas.append({
                    'D√≠a pedido': dias[t-1],
                    'D√≠a entrega': dias[t+1],
                    'Pedidos': pedidos_dia,
                    'Costo de Env√≠o': f"${pyo.value(modelo.k):.2f}"
                })
    
    if datos_entregas:
        for entrega in datos_entregas:
            print(f"{entrega['D√≠a pedido']} (llega el {entrega['D√≠a entrega']}): {entrega['Pedidos']} (Costo: {entrega['Costo de Env√≠o']})")
    else:
        print("¬°No hay entregas programadas!")
    
    # Niveles de inventario diarios
    print("\nüì¶ NIVELES DE INVENTARIO DIARIOS:")
    print("-" * 60)
    
    datos_inventario = []
    for t in modelo.T:
        fila = {'D√≠a': dias[t-1]}
        inventario_total = 0
        volumen_total = 0

        for i in modelo.I:
            nivel_inventario = pyo.value(modelo.s[i, t])
            volumen_inventario = pyo.value(modelo.v[i]) * pyo.value(modelo.s[i, t])
            fila[i] = f"{nivel_inventario:.1f}"
            inventario_total += nivel_inventario
            volumen_total += volumen_inventario
        fila['Total Unidades'] = f"{inventario_total:.1f}"
        fila['Volumen Total'] = f"{volumen_total:.1f}"
        datos_inventario.append(fila)
    
    df_inventario = pd.DataFrame(datos_inventario)
    print(df_inventario.to_string(index=False))
    
    # Desglose de costos
    print("\nüí∏ DESGLOSE DE COSTOS:")
    print("-" * 60)
    
    costo_ingredientes = sum(pyo.value(modelo.c[i]) * pyo.value(modelo.q[i, p]) * pyo.value(modelo.x[p, t]) 
                           for i in modelo.I for p in modelo.P for t in modelo.T)
    
    costo_entrega = sum(pyo.value(modelo.k) * pyo.value(modelo.z[t]) for t in modelo.T)

    costo_almacenamiento = sum(pyo.value(modelo.f) * pyo.value(modelo.v[i]) * pyo.value(modelo.s[i,t]) 
                              for i in modelo.I for t in modelo.T)

    costo_demanda_no_satisfecha_ve = sum(
        pyo.value(modelo.lambda_penalizacion) * pyo.value(modelo.pi[e]) * pyo.value(modelo.u[p,t,e]) 
        for p in modelo.P for t in modelo.T for e in modelo.E
    )

    print(f"Costos de Ingredientes: ${costo_ingredientes:.2f}")
    print(f"Costos de Entrega: ${costo_entrega:.2f}")
    print(f"Costos de Almacenamiento: ${costo_almacenamiento:.2f}")
    print(f"Costo Total: ${costo_ingredientes + costo_entrega + costo_almacenamiento:.2f}")
    print(f"Costo de Demanda No Satisfecha: ${costo_demanda_no_satisfecha_ve:.2f}")

# =============================================================================
# EJECUCI√ìN PRINCIPAL
# =============================================================================

if __name__ == "__main__":
    print("üçî DESAF√çO DE OPTIMIZACI√ìN HOUSE BURGER")
    print("=" * 60)
    print("Usando Programaci√≥n Lineal Entera Mixta (MILP) para optimizar:")
    print("‚Ä¢ Ventas diarias de productos")
    print("‚Ä¢ Pedidos de ingredientes")
    print("‚Ä¢ Programaci√≥n de entregas")
    print("‚Ä¢ Minimizaci√≥n de costos")
    print("=" * 60)
    
    # Resolver el problema de optimizaci√≥n
    modelo, resultados = resolver_optimizacion_house_burger()
    
    if modelo is not None:
        print("\nüéâ ¬°OPTIMIZACI√ìN COMPLETADA!")
        print("¬°Tu House Burger est√° ahora √≥ptimamente planificada para la semana!")
    else:
        print("\n‚ùå ¬°OPTIMIZACI√ìN FALL√ì!")
        print("Por favor verifica la formulaci√≥n del problema y las restricciones.")

# =============================================================================
# FUNCIONES DE AN√ÅLISIS ADICIONALES
# =============================================================================

def analisis_sensibilidad(modelo_base):
    """
    Realizar an√°lisis de sensibilidad en par√°metros clave
    """
    print("\nüîç AN√ÅLISIS DE SENSIBILIDAD:")
    print("-" * 60)
    
    # Analizar sensibilidad del costo de env√≠o
    costos_envio = [20, 25, 30, 35]
    print("Sensibilidad del Costo de Env√≠o:")
    
    for costo in costos_envio:
        # Crear nuevo modelo con diferente costo de env√≠o
        modelo_temp = modelo_base.clone()
        modelo_temp.k.set_value(costo)
        
        solver = pyo.SolverFactory("gurobi", solver_io="python")
        resultados = solver.solve(modelo_temp, tee=False)
        
        if resultados.solver.termination_condition == TerminationCondition.optimal:
            costo_total = pyo.value(modelo_temp.obj)
            print(f"  Env√≠o = ${costo}: Costo Total = ${costo_total:.2f}")

def exportar_resultados_a_excel(modelo, nombre_archivo="resultados_house_burger.xlsx"):
    """
    Exportar resultados de optimizaci√≥n a archivo Excel
    """
    try:
        with pd.ExcelWriter(nombre_archivo, engine='openpyxl') as writer:
            
            # Datos de ventas de productos
            dias = ['Lunes', 'Martes', 'Mi√©rcoles', 'Jueves', 'Viernes', 'S√°bado', 'Domingo']
            datos_ventas = []
            for t in modelo.T:
                fila = {'D√≠a': dias[t-1]}
                for p in modelo.P:
                    fila[p] = int(pyo.value(modelo.x[p, t]))
                datos_ventas.append(fila)
            
            df_ventas = pd.DataFrame(datos_ventas)
            df_ventas.to_excel(writer, sheet_name='Ventas de Productos', index=False)
            
            # Datos de inventario
            datos_inventario = []
            for t in modelo.T:
                fila = {'D√≠a': dias[t-1]}
                for i in modelo.I:
                    fila[i] = pyo.value(modelo.s[i, t])
                datos_inventario.append(fila)
            
            df_inventario = pd.DataFrame(datos_inventario)
            df_inventario.to_excel(writer, sheet_name='Niveles de Inventario', index=False)
            
            print(f"‚úÖ Resultados exportados a {nombre_archivo}")
    
    except ImportError:
        print("‚ö†Ô∏è openpyxl no disponible. Instala con: pip install openpyxl")
    except Exception as e:
        print(f"‚ùå Error exportando a Excel: {e}")

# Ejemplo de uso con an√°lisis de sensibilidad
if __name__ == "__main__":    
    # Resolver el problema de optimizaci√≥n
    modelo, resultados = resolver_optimizacion_house_burger()
    
    if modelo is not None:
        # Realizar an√°lisis de sensibilidad
        analisis_sensibilidad(modelo)
        
        # Exportar resultados
        exportar_resultados_a_excel(modelo)
        
        print("\nüéâ ¬°AN√ÅLISIS DE SENSIBILIDAD COMPLETADO!")
        print("¬°Tu House Burger est√° ahora √≥ptimamente planificada para la semana!")
    else:
        print("\n‚ùå ¬°AN√ÅLISIS DE SENSIBILIDAD FALL√ì!")
        print("Por favor verifica la formulaci√≥n del problema y las restricciones.")
```
